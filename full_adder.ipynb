```python
# Import Libraries
import pandas as pd
```

```python
# Integer to binary function
def binary(num):
    converted = []
    while (num):
        converted.append(1 if num % 2 else 0)
        num //= 2
    converted.reverse()
    return converted
```

```python
# Binary to integer function
def decimal(b):
    ret = 0
    base = 0
    b.reverse()
    for i in b:
        ret += 2**(base)*i
        base += 1
    return ret
```

```python
#Appends '0' to begenning of X and Y binary list until they are equal
def length_checker(x,y,z):
    x.reverse()
    y.reverse()
    z.reverse()
    m=max(len(x),len(y),len(z))
    while(len(x)<m):
        x.append(0)
    while(len(y)<m):
        y.append(0)
    while(len(z) < m):
        z.append(0)
    x.append(0)
    y.append(0)
    z.append(0)
    x.reverse()
    y.reverse()
    z.reverse()
    return x,y,z
```

```python
#Full adder function
def fullAdder(x_int,y_int,z_int):
    X_bin=binary(x_int)
    Y_bin=binary(y_int)
    Z_bin=binary(z_int)
    X_bin, Y_bin, Z_bin = length_checker(X_bin, Y_bin, Z_bin)
    
    sum = []
    carry = []
    #Sum = X xor Y xor Z #Carry = X and Y or Z and (X xor Y)
    for i in range(len(X_bin)):
        sum.append(X_bin[i] ^ Y_bin[i] ^ Z_bin[i])
        carry.append((X_bin[i] and Y_bin[i]) or ((X_bin[i]^Y_bin[i]) and Z_bin[i]))
    
    print(f"Input X : {x_int}")
    print(f"Input Y : {y_int}")
    print(f"Input Z : {z_int}\n")
    data=pd.DataFrame({"X binary":X_bin,"Y binary":Y_bin,"Z binary":Z_bin,"Sum":sum,"Carry":carry})
    data.index.name="S no"
    return data
```

```python
#X=15 Y=51 Z=85
fullAdder(15,51,85)
```
