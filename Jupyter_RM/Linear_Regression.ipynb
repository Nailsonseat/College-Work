```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([3.6,4.8,7.2,6.9,10.7,6.1,7.9,9.5,5.4]).reshape(-1,1).T
y=np.array([9.3,10.2,11.5,12,18.6,13.2,10.8,22.7,12.7]).reshape(-1,1)
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.figure(figsize=(10,10))
plt.xlim(-12,12)
plt.ylim(-23,23)
plt.axhline(color='black',linestyle='dashed',lw=2)
plt.axvline(color='black',linestyle='dashed',lw=2)
plt.scatter(X,y)
```

```python
dummy=np.ones(X.shape[1])
X_s=np.row_stack((dummy,X))
```

```python
def weights(X,y):
    return np.linalg.pinv(X@X.T)@X@y
def predict(X,w):
    return X.T@w
```

```python
w=weights(X_s,y)
print(w)
```

```python
plt.figure(figsize=(10,10))
plt.xlim(-12,12)
plt.ylim(-23,23)
s=np.linspace(-15,15,100)
t=w[0,0]+s*w[1,0]
plt.plot(s,t,c='red')
plt.axhline(color='black',linestyle='dashed',lw=2)
plt.axvline(color='black',linestyle='dashed',lw=2)
plt.scatter(X,y)
```
